#Theory Questions

1. What is the difference between interpreted and compiled languages?

- >Interpreted Languages:

  Code is executed line-by-line using an interpreter.

  Slower compared to compiled languages as each line is processed at runtime.

Example languages :  Python, JavaScript, Ruby.


- >Compiled Languages:

   Code is converted into machine code by a compiler before execution.

   Faster execution as the machine code is directly executed by the CPU.

Example languages : C, C++, Java (compiled into bytecode, then interpreted by JVM).

2. What is exception handling in Python?

  Exception handling is a mechanism to handle runtime errors in a program without terminating it unexpectedly.

  In Python, exceptions are handled using:

  try: Defines a block of code to monitor for errors.

  except: Handles the exception if it occurs.

  else: Executes if no exception occurs.

  finally: Executes regardless of whether an exception occurs (used for cleanup).


3. What is the purpose of the finally block in exception handling?

 The finally block is used to define code that will always execute, regardless of whether an exception occurred in the try block.
 Commonly used for cleanup tasks, such as closing files or releasing resources.

9. What is the difference between multithreading and multiprocessing?

- >Multithreading:

 Multiple threads within the same process share the same memory space.

 Useful for I/O-bound tasks (e.g., file reading, web scraping).


- >Multiprocessing:

 Each process has its own memory space, making it ideal for CPU-bound tasks (e.g., computations).
 Avoids the Global Interpreter Lock (GIL) in Python.

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=1_000_000, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.DEBUG)
logging.info("This is a log message.")

```

21.Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
word_to_find = "example"
with open("example.txt", "r") as file:
    content = file.read()
    count = content.count(word_to_find)
    print(f"The word '{word_to_find}' occurs {count} times.")

```

22. How can you check if a file is empty before attempting to read its contents.

```python
import os
if os.path.exists("example.txt") and os.stat("example.txt").st_size > 0:
    with open("example.txt", "r") as file:
        print(file.read())
else:
    print("The file is either empty or does not exist.")

```
